Vectorizando la data

Importando de sklearn una libreria la cual cuenta con la tecnica TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

Teniendo en cuenta que las palabras comunes tienen menos IDF y las palabras mas inusuales o unicas tienen mayor IDF, pasamos hiperparametros para no tener en cuenta las palabras no utiles o las palabras mas comunes

Creamos un dataframe con los datos convertidos a vectores

## Reduciendo la dimension del dataframe

A traves de la libreria TruncatedSVD podemos reducir la dimension del dataframe

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=3000)

reduced_data = svd.fit_transform(vectorized_dataframe)
```

## Ya que tenemos todo el dataframe limpio y en su mejor estado, pasamos a realizar las recomendaciones

Implementaremos la formula algebraica de similitud de cosenos para realizar las recomendaciones

```python
from sklearn.metrics.pairwise import cosine_similarity
```

Realizamso la similitud de cosenos utilizando la variable que contiene toda la data reducida

```python
similarity = cosine_similarity(reduced_data)
```

```python
similarity
```

Finalmente creamos una funcion la cual toma como parametro el nombre de una pelica que exista en el dataframe. Luego como primera instacia, busca en el dataframe, en la columna title, si el el parametro coincide con algun nombre del dataframe, si es asi, se recomenda primero a las peliculas las cuales contengan el mismo nombre, en el caso contrario, hacemos una variable que se llame "distances" la cual contiene las similitudes de las diferentes pelicas contenidas en un array. Luego hacemos una lista la cual tendra 15 elementos. Esta misma tendra de forma enumerada las distancias obtenidas en la variable declarada anteriormente. Finalemente, mostramos con un for el cual recorre la lsita recien creada los resultados obtenidos

```python
def recomendation_system(movie):
    id_of_movie = movies_df[movies_df['title']==movie].index[0]
    distances = similarity[id_of_movie]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:15]
    for movie_id in movie_list:
        print(movies_df.iloc[movie_id[0]].title)
```

```python
recomendation_system('The Matrix')
```

```python
recomendation_system('Jumanji')
```

```python
recomendation_system('Casino')
```
